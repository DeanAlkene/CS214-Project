## Load Table-A

```python
import csv
import pandas as pd
import numpy as np
```

```python
orders_csv = pd.read_csv("../DATA-PreProcess/sourceData/TableA-Orders.csv")
orders_csv
```

## Display Table-A

```python
orders_df = pd.DataFrame(orders_csv)
orders_df = orders_df.sort_values(
    by=["City of Seller", "City of Purchaser", "Time of Order"])
orders_df
```

## transform Table-A (str -> float)

```python
def Eval(Str):
    if type(Str)!=type(""):
        return Str
    Lstrip = Str.lstrip('0')
    if Lstrip == "":
        return 0
    else:
        return eval(Lstrip)

def Time(Str): # 09:45:12 -> float (min)
    Split = Str.split(':')
    return Eval(Split[0]) * 60 + Eval(Split[1]) + Eval(Split[2]) / 60

Time("00:45:12")
```

```python
orders_df['Time of Order'] = orders_df['Time of Order'].map(Time)
```

```python
orders_df.to_csv("orders.csv", index=None)
```
